```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx

df = pd.read_csv("dataset.csv")
df.head()
```

```python
df.replace("?", np.nan, inplace=True)
```

```python
# df2=df["age"]
# df3=df["connections"]
```

```python
graph = nx.Graph()

graph = nx.from_pandas_edgelist(df, source="source", target="connections")
```

```python
graph.nodes()
```

```python
options = {
    "font_size": 10,
    "node_size": 500,
    "node_color": "white",
    "edgecolors": "green",
    "linewidths": 4,
    "width": 4,
}
```

```python
# nx.draw(graph)
nx.draw_networkx(graph, **options)
```
